```python
# Print a header for the shop menu
print("Welcome to the shop!")
print("Here are the items and prices:")

# Define the store inventory
inventory = {
    'item1': {'name': 'Biscuits', 'quantity': 50, 'price': 10},
    'item2': {'name': 'Oats Meal', 'quantity': 30, 'price': 45},
    'item3': {'name': 'Dry Fruits', 'quantity': 20, 'price': 42},
    'item4': {'name': 'Lentils', 'quantity': 20, 'price': 30},
    'item5': {'name': 'Rice', 'quantity': 20, 'price': 25}
}

# Create an empty list to store the customer's order.
customer_order = []

# Function to get an item from the inventory and add it to the customer's order
def add_item_to_order(item_key, quantity):
    item_info = inventory[item_key]
    item_name = item_info["name"]
    
    if quantity <= item_info["quantity"]:
        cost_per_item = item_info["price"]
        customer_order.append({"name": item_name, "quantity": quantity, "price": cost_per_item})
        # Update the inventory to reflect the purchase
        item_info["quantity"] -= quantity
    else:
        print(f"Sorry, we don't have enough {item_name} in stock.")

# Function to sell items to customers and update inventory
def sell_item(item_name, quantity):
    if item_name in inventory and inventory[item_name]["quantity"] >= quantity:
        item_info = inventory[item_name]
        item_info["quantity"] -= quantity
        print(f'Sold {quantity} {item_name}(s). Remaining stock: {item_info["quantity"]}')
    else:
        print(f'Sorry, insufficient stock for {item_name}.')

# Ask the customer for their order
while True:
    while True:
        # Display the available items
        print("\nAvailable items:")
        for item_key, item_info in inventory.items():
            print(f"{item_info['name']} - ${item_info['price']} ({item_info['quantity']} in stock)")

        # Ask the customer for their order
        item_key = input("Enter the item number you want to buy (e.g.,item1) : ")
        quantity = int(input(f"How many {inventory[item_key]['name']} do you want to buy? "))

        # Add the selected item to the customer's order
        add_item_to_order(item_key, quantity)

        # Prompt the user to continue shopping or check out
        response = input("Do you want to continue shopping? (yes/no): ")
        if response.lower() != "yes":
            break

    # Define a function to get customer information
    def get_customer_info():
        customer_name = input("What is your name? ")
        delivery_address = input("What is your delivery address? ")
        customer_phone = input("What is your phone number? ")
        # Return the customer's information as a dictionary
        return {
            "customer_name": customer_name,
            "address": delivery_address,
            "phone": customer_phone,
        }

    # Get the customer's information
    customer_details = get_customer_info()

    # Function to calculate delivery charges based on distance
    def delivery_charges(distance):
        if distance <= 5:
            return 10
        elif 5 < distance <= 10:
            return 12
        elif 10 < distance <= 15:
            return 14
        else:
            return 20

    # Get the distance from the user
    distance = float(input("Enter the distance in miles: "))

    # Calculate delivery charges
    charges = delivery_charges(distance)

    # Print the customer's name, address, and delivery charges
    print(f"Customer Name: {customer_details['customer_name']}")
    print(f"Customer Address: {customer_details['address']}")
    print(f"Customer Phone: {customer_details['phone']}")
    print(f"Delivery Charges: ${charges}")

    # Function to print the final bill
    def print_bill(order_items, customer_info, delivery_charge):
        # Print the header
        print("-" * 50)
        print("Final Bill")
        print("-" * 50)
        # Print the customer details    print("Customer Name:", customer_info["customer_name"])
        print("Customer Address:", customer_info["address"])
        print("Customer Phone:", customer_info["phone"])
        # Print the items purchased
        print("Items Purchased:")
        total_cost = 0
        for item in order_items:
            item_total = item["quantity"] * item["price"]
            total_cost += item_total
            print(f"{item['name']}: {item['quantity']} x ${item['price']} = ${item_total}")
        # Print the delivery charge
        print(f"Delivery Charge: ${delivery_charge}")
        # Calculate and print the total bill
        total_bill = total_cost + delivery_charge
        print(f"Total Bill: ${total_bill}")

    # Print the final bill
    print_bill(customer_order, customer_details, charges)

    # Prompt the user to continue shopping or check out
    response = input("Do you want to continue shopping? (yes/no): ")
    if response.lower() != "yes":
        break
    
# Print updated inventory after sales
print("\nUpdated Inventory:")
for item_key, item_info in inventory.items():
    print(f'{item_info["name"]}: {item_info["quantity"]} in stock')

    
print("\nThank you for shopping with us!")
```
